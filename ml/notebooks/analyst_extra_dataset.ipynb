```python
import os
import pandas as pd
```

```python
data_source = {
    "drug_drug_little" : r"C:\GitHub Project\dda - backend\data\raw\drug vs drug\db_drug_interactions.csv",
    "drug_drug_big" : r"C:\GitHub Project\dda - backend\data\raw\drug vs drug\twosides_with_drug_names.csv"
}
```

```python
for name, file_path in data_source.items():
    print(f"--------{name}--------")
    df = pd.read_csv(file_path)
    print('Shape ', df.shape)
    print('Columns:', df.columns.tolist())
    print()
    print(df.head(5).to_string())
    print('Total pairs:', len(df))
    print('Unique Drug 1:', df['Drug 1'].nunique())
    print('Unique Drug 2:', df['Drug 2'].nunique())

    all_drugs = pd.concat([df['Drug 1'], df['Drug 2']]).unique()
    print('Total unique drugs:', len(all_drugs))

    print()
    print('Sample interactions:')
    print(df.sample(5)[['Drug 1', 'Drug 2', 'Interaction Description']].to_string())
    print("\n" + "="*40 + "\n")
```
